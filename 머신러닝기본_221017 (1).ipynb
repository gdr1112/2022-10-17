```python
fish_data = [[l, w] for l, w in zip(length, weight)]

print(fish_data)
```

- 지도학습은 라벨이 필요하다. 
--> 종속변수, 타겟변수 (Response Variable, Dependent Variable)

```python
# 도미 35개 1로 표시
# 빙어 14개 0로 표시
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

## 머신러닝 모형 구현
- KNN 구현

```python
from sklearn.neighbors import KNeighborsClassifier 

# 모형 불러오기
kn = KNeighborsClassifier()

# 모형 학습
kn.fit(fish_data, fish_target)
```

```python
# 모형 평가
# 1.0의미 ==> 100% 정확도
kn.score(fish_data, fish_target)
```

```python
# 예측 (= 배포)
new_data = [[30, 600]]
kn.predict(new_data) # 도미로 예측함
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(bream_length, bream_weight)
ax.scatter(smelt_length, smelt_weight)
ax.scatter(30, 600, marker = '^')
ax.set_xlabel('length')
ax.set_ylabel('weight')

plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier 

# 모형 불러오기
kn = KNeighborsClassifier(n_neighbors = 17)

# 모형 학습
kn.fit(fish_data, fish_target)

# 모형 평가
# 0.71의미 ==> 71% 정확도
kn.score(fish_data, fish_target)
```

```python
for n in range(5, 50):
  # 최근접 이웃 개수 설정
  kn.n_neighbors = n

  # 스코어 계산
  score = kn.score(fish_data, fish_target)

  # 1.0에 미치지 못하는 이웃 개수 경계선 구하기
  if score < 1:
    print(n, score)
    break
```

## 머신러닝 중요 메서드
- fit() : 훈련시 사용하는 메서드, 두개의 데이터가 들어감. 
  + 독립변수 : fish_data (길이, 몸무게)
  + 종속변수 : fish_target

- predict() : 예측할 때 사용
  + 새로운 데이터 : 독립변수만 추가

- score() : 모형의 성능 평가
  + 실전 : 평가지표 함수를 사용!

```python
a_list = [1,2,3]
b_list = [4,5,6]

np.column_stack((a_list, b_list))
```

```python
fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data[:5])
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
print(fish_target)
```

Scikit-Learn  라이브러리가 요구하는 input 데이터 형태는 기본적으로 배열임(리스트가 아님)

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state=42
)

print(train_input.shape, train_target.shape)
```

```python
print(test_input.shape, test_target.shape)
```

 - 모형만들기


```python
kn = KNeighborsClassifier(n_neighbors = 5)
kn= kn.fit(train_input, train_target)
kn.score(test_input , test_target)
```

## Feature Engineering 맛보기
  + 표준점수 (Z- Score)

```python
mean = np.mean(train_input ,axis = 0)  # 평균
std = np.std(train_input, axis = 0) #표준편차

print(mean, std)
```

```python
train_scaled = (train_input - mean) / std
train_scaled
```

```python
# 모델링 재학습
kn.fit(train_scaled, train_target)
```

```python
# 모형 평가 ==> 100%
test_scaled = (test_input - mean) /std
kn.score(test_scaled, test_target)
```

```python
# 새로운 데이터 ==> 표준점수화
new = ([25,150] -mean) /std
new
```

```python
print(kn.predict([new]))
```

```python
distances, indexes = kn.kneighbors([new])

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(train_scaled[:, 0], train_scaled[:, 1])
ax.scatter(new[0], new[1], marker = '^')
ax.scatter(train_input[indexes, 0], train_input[indexes, 1], marker='D')
ax.set_xlabel('length')
ax.set_ylabel('weight')

plt.show()
```
